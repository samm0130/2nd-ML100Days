- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
```

```python
for  i in range(2,8):
    output=KMeans(n_clusters=i).fit(X)
    print i
    print output.labels_
    
```

```python
for  i in range(2,8):
    output=KMeans(n_init=5,n_clusters=i).fit(X)
    print i
    print output.labels_
```
